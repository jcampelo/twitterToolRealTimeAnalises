```python
base = []
```

```python
for i in dados:
           
    linha = [ 0 for j in range (5)]
    linha[0] = i.text
    texto = i.text
    
    if ('RT' in texto):
        
        posicao = texto.find(':')
        texto = texto[posicao+2:]
        linha[4] = 1
        
    linha[1] = 1 if ('Bolsonaro' in texto.lower() or '22' in texto.lower()) else 0
    linha[2] = 1 if ('Lula' in texto.lower() or '13' in texto.lower()) else 0
    linha[3] = 1 if ('Ciro' in texto.lower() or '12' in texto.lower()) else 0
    
    base.append(linha)
        
    
    
```

```python
baseEleicao = pd.DataFrame(base)
baseEleicao.columns = ['texto','bolsonaro','lula','ciro','RT']
display(baseEleicao)
```

```python
baseView = baseEleicao.drop(['texto','RT'],axis=1)
```

```python
comentarios = baseView.sum().sort_values(ascending=False).to_list()
```

```python
participantes = baseView.sum().sort_values(ascending=False).index.to_list()
```

```python
base2 = pd.DataFrame(zip(participantes,comentarios))
base2.columns = ['pessoas','comentarios']
```

```python
total_comentarios = base2['comentarios'].sum()
```

```python
base2['perc'] = base2['comentarios']/total_comentarios
```

```python
base2['perc_form'] = base2['perc'].map("{:.1%}".format)
display(base2[['pessoas','comentarios','perc_form']])
```
